```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split as tts
from sklearn.preprocessing import MinMaxScaler


from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier


from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df = pd.read_csv('/content/obesity_data.csv')
df.head()
```

```python
df.corr()
```

```python
bin = {'Obese': 1, 'Overweight': 1, 'Normal weight': 0, 'Underweight': 0}
```

```python
bin2 = {'Male': 1, 'Female': 0}
```

```python
df['obese or not'] = df['ObesityCategory'].map(bin)
```

```python
df['Sex'] = df['Gender'].map(bin2)
```

```python
df.drop('Gender',axis=1,inplace=True)
```

```python
df.drop('ObesityCategory',axis=1,inplace=True)
```

```python
df.drop('BMI',axis=1,inplace=True)
```

```python
df.head()
```

```python
X = df.drop('obese or not', axis = 1)
y = df['obese or not']
```

```python
X_train, X_test, y_train, y_test = tts(X, y, test_size = 0.2, random_state = 123)
```

```python
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = RandomForestClassifier()
```

```python
from sklearn.metrics import accuracy_score

model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
test_data = pd.DataFrame({'Age': [20,65,71], "Weight":[90,48,72],'PhysicalActivityLevel': [1,3,2],"Sex":[1,0,1],})


test_data_scaled = scaler.transform(test_data)


prediction = model.predict(test_data_scaled)


print("Fat or not:", prediction)
for x in prediction:
  if (x==1):
      print("fat")

  else:
    print("not fat")
```
